# **Train the Model**

```python
model = DecisionTree(max_depth=5)
```

```python
X = iris.values.tolist()
y = []
for row in X:
    y.append(int(row[4]))
    del row[4]
X = pd.Series(X)
y = pd.Series(y)

#X_train, X_test = train_test_split(X, test_size=0.2, shuffle=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

X_train_list = X_train.values.tolist()
X_test_list = X_test.values.tolist()
y_train_list = y_train.values.tolist()
y_test_list = y_test.values.tolist()
```

```python
model.fit(X_train_list)
```

```python
train_predicted = model.predict(X_train_list)
test_predicted = model.predict(X_test_list)
```

```python
print('Expected Classification on Test Set')
print(y_test_list)
print('Predictions')
print(test_predicted)
print('Expected Classification on Train Set')
print(y_train_list)
print('Predictions')
print(train_predicted)
```

# **Results**

### **Confusion Matrix of Test Set**

```python
mt = metrics.confusion_matrix(y_test_list, test_predicted)
print(sns.heatmap(mt, annot=True))
```

### **Confusion Matrix of Train Set**

```python
mt = metrics.confusion_matrix(y_train_list, train_predicted)
print(sns.heatmap(mt, annot=True))
```

### **Accuracy**

```python
test_acc = metrics.accuracy_score(y_test_list, test_predicted)
print('Accuracy of Test')
print(test_acc)
train_acc = metrics.accuracy_score(y_train_list, train_predicted)
print('Accuracy of Train')
print(train_acc)
```

### **F1-Score**

```python
f1_test = metrics.f1_score(y_test_list, test_predicted, average='weighted')
print('F1-Score Test')
print(f1_test)

f1_train = metrics.f1_score(y_train_list, train_predicted, average='weighted')
print('F1-Score Train')
print(f1_train)
```

### **Precision**

```python
precision_test = metrics.precision_score(y_test_list, test_predicted, average='weighted')
print('Precision Score Test')
print(precision_test)

precision_train = metrics.precision_score(y_train_list, train_predicted, average='weighted')
print('Precision Score Train')
print(precision_train)
```

### **Recall**

```python
recall_test = metrics.recall_score(y_test_list, test_predicted, average='weighted')
print('Recall Score Test')
print(recall_test)

recall_train = metrics.recall_score(y_train_list, train_predicted, average='weighted')
print('Recall Score Train')
print(recall_train)
```

### **ROC/AUC Curve**

```python
y_pred2 = pd.Series(test_predicted)
y_test2 = pd.Series(y_test_list)

y_testb = label_binarize(y_test2, classes=[0,1,2])
y_predb = label_binarize(y_pred2, classes=[0,1,2])

fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(3):
    fpr[i], tpr[i], _ = roc_curve(y_testb[:,i], y_predb[:,i])
    roc_auc[i] = auc(fpr[i], tpr[i])
                                          
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(3)]))
mean_tpr = np.zeros_like(all_fpr)
for i in range(3):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
mean_tpr /= 3;
fpr["macro"]= all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

colors = cycle(['blue', 'yellow', 'red'])
for i, color in zip(range(3), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2)

plt.plot([0,1], [0,1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive')
plt.ylabel('True Positive Rate')
plt.show()
print(roc_auc["macro"])
```
